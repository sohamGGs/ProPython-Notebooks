```python
class InvalidAgeError(Exception):
    """Custom exception for age-restricted content."""
    pass

def register_user(username: str, age: int):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    if age < 18:
        raise InvalidAgeError(f"User {username} is too young for this platform.")

    print(f"User {username} registered successfully!")

if __name__ == "__main__":
    try:
        user_age = int(input("Enter your age: ")) # Colab supports input()
        register_user("Soham_Patil", user_age)
    except InvalidAgeError as e:
        print(f"Access Denied: {e}")
    except ValueError as e:
        print(f"Input Error: {e}")
```
