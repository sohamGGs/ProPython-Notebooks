```python
from typing import Set

# Day 4: Concept - Set Operations
list_a = ["python", "java", "c++", "python", "javascript"]
list_b = ["c++", "go", "rust", "python"]

# Deduplication
set_a: Set[str] = set(list_a)
set_b: Set[str] = set(list_b)

# Set Logic
intersection = set_a.intersection(set_b)  # Common in both
difference = set_a.difference(set_b)      # In A but not in B
symmetric_diff = set_a ^ set_b            # In either A or B, but not both

print(f"Common Skills: {intersection}")
print(f"Unique to A: {difference}")
print(f"Exclusive to either: {symmetric_diff}")
```
