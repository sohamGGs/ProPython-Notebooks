```python
from typing import Set, List

def get_unique_words(text: str) -> List[str]:
    """
    Cleans text and returns a sorted list of unique words.
    """
    # Remove basic punctuation and lowercase everything
    clean_text = "".join(char.lower() for char in text if char.isalnum() or char.isspace())

    # Using a set for instant deduplication
    words: Set[str] = set(clean_text.split())

    return sorted(list(words))

if __name__ == "__main__":
    sample_text = """
    Python is amazing. Python is fast. Learning Python in 50 days
    is a great goal for any Python developer!
    """
    unique = get_unique_words(sample_text)
    print(f"Total Unique Words: {len(unique)}")
    print(f"Word List: {unique}")
```
