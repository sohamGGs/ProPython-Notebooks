```python
def analyze_grades(grades: List[int]) -> Dict[str, Union[float, List[int]]]:
    """
    Uses comprehensions to filter and analyze a list of student grades.
    """
    passing_score = 40
    passed = [g for g in grades if g >= passing_score]
    failed = [g for g in grades if g < passing_score]

    return {
        "average": sum(grades) / len(grades) if grades else 0,
        "passed_count": len(passed),
        "top_performers": [g for g in grades if g >= 90],
        "fail_list": failed
    }

if __name__ == "__main__":
    class_grades = [88, 34, 92, 45, 21, 99, 67]
    stats = analyze_grades(class_grades)
    for key, value in stats.items():
        print(f"{key.replace('_', ' ').title()}: {value}")
```
