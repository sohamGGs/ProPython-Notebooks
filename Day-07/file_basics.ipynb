```python
# Day 7: The 'with' statement (Context Manager)
filename = "learning_log.txt"

# Writing to a file
with open(filename, "w") as file:
    file.write("Day 7: Learning File I/O in Python!\n")
    file.write("Context managers handle closing files automatically.")

# Reading from a file
with open(filename, "r") as file:
    content = file.read()
    print("--- File Content ---")
    print(content)
```
