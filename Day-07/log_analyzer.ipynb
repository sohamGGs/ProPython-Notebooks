```python
import os
from typing import List

def create_dummy_log(name: str):
    logs = [
        "INFO: System Started",
        "ERROR: Database Connection Failed",
        "INFO: User Login: Soham",
        "ERROR: Timeout on Request 404",
        "INFO: Backup Completed"
    ]
    with open(name, "w") as f:
        f.write("\n".join(logs))

def extract_errors(log_file: str) -> List[str]:
    errors = []
    if not os.path.exists(log_file):
        return ["File not found."]

    with open(log_file, "r") as file:
        for line in file:
            if "ERROR" in line:
                errors.append(line.strip())
    return errors
if __name__ == "__main__":
    log_name = "system.log"
    create_dummy_log(log_name)

    found_errors = extract_errors(log_name)
    print(f"Found {len(found_errors)} critical errors:")
    for err in found_errors:
        print(f"-> {err}")
```
