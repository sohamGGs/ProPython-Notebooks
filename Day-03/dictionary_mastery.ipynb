```python
from typing import Dict, Any

# Advanced Dictionary Operations
user_profile: Dict[str, Any] = {
    "username": "soham_patil",
    "role": "Developer",
    "tech_stack": ["Python", "Django", "DS"]
}

# 1. Using .get() to avoid KeyErrors
location = user_profile.get("location", "Not Specified")

# 2. Dictionary Comprehension (Filtering)
scores = {"Math": 95, "Physics": 88, "Chemistry": 72, "English": 45}
distinction_subjects = {k: v for k, v in scores.items() if v >= 75}

# 3. Merging Dictionaries (Python 3.9+)
update_data = {"location": "Pune", "role": "Senior Developer"}
user_profile |= update_data

print(f"Updated Profile: {user_profile}")
print(f"Distinction In: {distinction_subjects}")
```
