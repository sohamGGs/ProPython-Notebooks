```python
from typing import Dict, Optional

class ContactBook:
    def __init__(self):
        # Dictionary to store name as key and phone as value
        self.contacts: Dict[str, str] = {}

    def add_contact(self, name: str, phone: str) -> str:
        self.contacts[name] = phone
        return f"Contact '{name}' added successfully."

    def get_contact(self, name: str) -> str:
        return self.contacts.get(name, "Contact not found.")

    def delete_contact(self, name: str) -> str:
        if name in self.contacts:
            del self.contacts[name]
            return f"Contact '{name}' deleted."
        return "Contact not found."

def run_app():
    book = ContactBook()
    print("--- Welcome to Colab Contact Book ---")

    # Simulating a quick session
    print(book.add_contact("Soham", "9876543210"))
    print(book.add_contact("Alice", "1234567890"))

    # Search simulation
    search_name = "Soham"
    print(f"Searching for {search_name}: {book.get_contact(search_name)}")

    print("\nCurrent Contacts:", book.contacts)

if __name__ == "__main__":
    run_app()
```
