```python
from typing import List

# Sample Data: Prices in INR
prices = [1200, 450, 900, 3200, 150, 2000]
# 1. Lambda: Anonymous functions
square = lambda x: x**2
# 2. Filter: Keep only prices > 1000
expensive_items = list(filter(lambda p: p > 1000, prices))
# 3. Map: Apply 18% GST (Tax) to all prices
prices_with_tax = list(map(lambda p: round(p * 1.18, 2), prices))

print(f"Original: {prices}")
print(f"Expensive (>1000): {expensive_items}")
print(f"With 18% GST: {prices_with_tax}")
```
