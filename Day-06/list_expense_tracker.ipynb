```python
from typing import List, Dict, Tuple

def analyze_expenses(expenses: List[Tuple[str, float, str]]) -> None:

    food_total = sum(map(lambda x: x[1], filter(lambda x: x[2] == "Food", expenses)))
    item_names = list(map(lambda x: x[0].upper(), expenses))

    print(f"All Items: {', '.join(item_names)}")
    print(f"Total Spent on Food: ₹{food_total}")

if __name__ == "__main__":
    my_expenses = [
        ("Lunch", 250.0, "Food"),
        ("Bus Ticket", 50.0, "Travel"),
        ("Dinner", 400.0, "Food"),
        ("Notebook", 100.0, "Education")
    ]
    analyze_expenses(my_expenses)
```
