```python
from typing import Optional
```

```python
def convert_length(value: float, from_unit: str, to_unit: str) -> Optional[float]:
    """
    Converts units between Meters, Kilometers, and Miles.
    """
    # Conversion factors to Meters
    to_meters = {"m": 1.0, "km": 1000.0, "mi": 1609.34}

    if from_unit not in to_meters or to_unit not in to_meters:
        return None

    # Convert to base unit (meters) then to target
    meters = value * to_meters[from_unit]
    return meters / to_meters[to_unit]

if __name__ == "__main__":
    val = 10
    start, end = "mi", "km"
    converted = convert_length(val, start, end)

    if converted is not None:
        print(f"{val}{start} is approximately {converted:.2f}{end}")
    else:
        print("Invalid units provided.")
```
