```python
from typing import List, Dict, Union, Optional

def process_user_data(name: str, age: int, scores: List[float]) -> Dict[str, Union[str, float]]:
    """
    Demonstrates type hinting for primitive types and collections.
    """
    average_score = sum(scores) / len(scores) if scores else 0.0
    return {
        "user": name,
        "status": "Adult" if age >= 18 else "Minor",
        "avg": average_score
    }

if __name__ == "__main__":
    result = process_user_data("Soham", 21, [85.5, 90.0, 78.5])
    print(f"Processed Data: {result}")
```
