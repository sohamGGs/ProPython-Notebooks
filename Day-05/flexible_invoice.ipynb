```python
from typing import Dict, Union

def generate_invoice(customer: str, *items: str, **costs: float) -> None:
    """
    Generates a simple text-based invoice.
    """
    print(f"\n--- INVOICE FOR: {customer.upper()} ---")
    total = 0.0

    for item in items:
        price = costs.get(item, 0.0)
        print(f"{item.capitalize()}: ${price:.2f}")
        total += price

    print(f"---------------------------")
    print(f"TOTAL DUE: ${total:.2f}")

if __name__ == "__main__":
    # Passing item names as *args and prices as **kwargs
    generate_invoice(
        "Soham Patil",
        "laptop", "mouse", "keyboard",
        laptop=1200.00,
        mouse=25.50,
        keyboard=45.00
    )
```
