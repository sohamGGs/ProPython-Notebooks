```python
from typing import Any

def debug_logger(message: str, *args: Any, **kwargs: Any) -> None:
    """
    A function that mimics a flexible logging system.
    """
    print(f"LOG: {message}")

    if args:
        print(f"Additional Info: {args}")

    if kwargs:
        print(f"Metadata: {kwargs}")

# Test calls
debug_logger("System Update")
debug_logger("User Login", "Soham", "192.168.1.1", status="Success", attempts=1)
```
